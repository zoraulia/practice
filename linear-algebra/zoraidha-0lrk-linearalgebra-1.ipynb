## **Toko Pak Brutus**

- Pak Brutus baru saja membuka toko online. Untuk sekarang, barang jualannya ada 3 item:
   - Sepatu
   - Tas
   - Donat

- Ferio dan Adam jadi 2 orang beruntung yang jadi pembeli pertama toko pak Brutus.
   - Ferio membeli 2 sepatu, 1 tas, dan 10 donat.
   - Adam membeli 5 sepatu dan 20 donat.

- Pak Brutus ingin melakukan beberapa perhitungan terkait jumlah pemasukan dan keuntungan yang didapat.
- Karena pak Brutus sudah belajar `Python` dan `Linear Algebra`, maka pak Brutus ingin membuat perhitungan menggunakan secara vektor.

---
**Soal 6**

- Buat vektor (menggunakan `NumPy`) yang mendeskripsikan total belanjaan Ferio & Adam
- Urutan elemen dalam vektor: `[jumlah sepatu, jumlah tas, jumlah donat]`
- Simpan vektor dalam variabel `item_ferio` dan `item_adam`

```python
item_ferio = np.array([2,1,10])
item_adam = np.array([5,0,20])
```

---
**Soal 7**

- Apabila harga
   - sepatu = Rp 200.000
   - tas = Rp 150.000
   - donat = Rp 50.000
- Berapa total belanjaan ferio & adam?
- Simpan dalam variabel `total_ferio` dan `total_adam`
- Lakukan perhitungan perkalian skalar dan vektor

```python
#perhitungan skalar
total_ferio = 200000*2 + 150000*1 + 50000*10
total_adam = 200000*5 + 50000*20
print(total_ferio)
print(total_adam)
```

```python
#perhitungan vektor
harga = np.array([200000, 150000, 50000])
total_ferio = np.dot(harga, item_ferio)
total_adam = np.dot(harga, item_adam)
print(total_ferio)
print(total_adam)
```

**Soal 8**

- Berapa total item toko online yang telah dibeli?
- Simpan dalam variabel `item_beli`

```python
item_beli = np.sum(np.add(item_ferio,item_adam))
item_beli
```

**Soal 9**

- Apabila modal untuk tiap item adalah sebagai berikut:
   - Sepatu = Rp 150000
   - Tas = Rp 100000
   - Donat = Rp 5000
- Berapakah untung yang didapat pak Brutus dari dua pembeli pertamanya?
- Simpan dalam variabel `total_untung`
- Lakukan perhitungan perkalian skalar dan vektor

```python
modalarr = np.array([150000, 100000, 5000])
ferio_modal = np.dot(modalarr, item_ferio)
adam_modal = np.dot(modalarr, item_adam)
total_modal = np.add(ferio_modal, adam_modal)
total_beli = total_ferio+total_adam
total_untung = total_beli-total_modal
total_untung
```
