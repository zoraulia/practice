**Menggunakan Library sklearn**

kita dapat menggunakan library clustering yang disediakan sklearn untuk melakukan kMeans algorithm.

```python
from sklearn.cluster import KMeans
```

```python
# point
points = np.concatenate([[0.3*np.random.randn(2) for i in range(100)],
                        [[1,1] + 0.3*np.random.randn(2) for i in range(100)], 
                        [[1,-1] + 0.3*np.random.randn(2) for i in range(100)]])
```

```python
# Plot Initial
plt.scatter(points[:,0], points[:,1], c='grey', s=10, label='point')
plt.xlim(-1.0, 2.1)
plt.ylim(-2.1, 2.1)
plt.legend()
plt.show()
```

Jalankan k-means algorithm

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(points)

c = kmeans.labels_
centroids = kmeans.cluster_centers_
J_obj_list = kmeans.inertia_
```

```python
# Plot Final
for i in range(len(c)):
    if c[i] == 1:
        plt.scatter(points[i,0], points[i,1], c='r', marker='+')
    elif c[i] == 2:
        plt.scatter(points[i,0], points[i,1], c='g', marker='+')
    else:
        plt.scatter(points[i,0], points[i,1], c='b', marker='+')
plt.scatter(centroids[0,0], centroids[0,1], c='r', edgecolor='k', s=100, label='centroid 1')
plt.scatter(centroids[1,0], centroids[1,1], c='g', edgecolor='k', s=100, label='centroid 2')
plt.scatter(centroids[2,0], centroids[2,1], c='b', edgecolor='k', s=100, label='centroid 3')

plt.xlim(-1.0, 2.1)
plt.ylim(-2.1, 2.1)
plt.legend()
plt.show()

print(f'centroids:')
print(f'centroid-{1} : {np.round(centroids[0], 2)}')
print(f'centroid-{2} : {np.round(centroids[1], 2)}')
print(f'centroid-{3} : {np.round(centroids[2], 2)}')
```
