**Soal 7**

- Pertama, buat vektor kategori post bagus, post biasa, post kurang, dan post Ferio dengan parameter post seperti pada soal. (Parameter berurutan: jumlah likes, shares, dan saved)

```python
post_bagus = np.array([50, 75, 50])
post_biasa = np.array([250, 10, 20])
post_kurang = np.array([100, 5, 10])
post_ferio = np.array([50, 25, 27])
```

---
**Soal 8**

- Cari jarak (Euclidean distance) antara post Ferio dengan masing-masing kategori post

```python
jarak_bagus = np.linalg.norm(post_ferio-post_bagus)
jarak_biasa = np.linalg.norm(post_ferio-post_biasa)
jarak_kurang = np.linalg.norm(post_ferio-post_kurang)

print(f"jarak_bagus: {jarak_bagus}")
print(f"jarak_biasa: {jarak_biasa}")
print(f"jarak_kurang: {jarak_kurang}")
```
