# 1.
Tuan Mor memiliki sebuah array dengan panjang N. Dia ingin menghitung jumlah selang-seling dari array tersebut. Jumlah selang-seling dari suatu array a adalah a1 − a2 + a3 − a4 + ..<br>
image.png<br>
Penjelasan: 1-4+2-5+5 = -1.<br>
image.png

```python
n = int(input('Masukkan panjang array: '))
array = []
for i in range(n):
  elemen = int(input(f'Masukkan elemen array ke {i+1}: '))
  array.append(elemen)

jumlah = 0
for j in range(n):
  if j%2 == 0:
    jumlah+=array[j]
  else:
    jumlah-=array[j]
print(f"jumlah selang-selingnya adalah {jumlah}")
```
