    Tuliskan biaya yang harus dibayar Lala sebagai operasi di bawah ini dan jalankan perhitungannya.

| barang      | harga  |
|-------------|--------|
| ayam        | 20000  |
| ikan kembung| 15000 |
| sayur kangkung | 10000  |
| susu        | 22000  |

```python
def payment():
    ayam = input("Jumlah ayam yang dibeli:")
    ikan = input("Jumlah ikan yang dibeli:")
    kangkung = input("Jumlah kangkung yang dibeli:")
    susu = input("Jumlah susu yang dibeli:")

    biaya = int(ayam)*20000 + int(ikan)*150000 + int(kangkung)*10000 + int(susu)*22000
    diskon = biaya*0.25
    total = biaya-diskon

    return total

jawaban_12 = payment()
print(jawaban_12)
```
